```python
df = df.drop(['date_of_admission', 'date_of_discharge'], axis=1)
```

```python
(df == "EMPTY").sum().sum()
```

```python
df.replace("EMPTY", 0, inplace=True)
```

```python
# Define lab columns
lab_cols = [
    'hemoglobin',
    'total_leukocyte_count',
    'platelet_count',
    'glucose_level',
    'urea_level',
    'creatinine_level'
]

# Create mask for invalid lab values (NA or 0)
invalid_mask = (df[lab_cols].isna()) | (df[lab_cols] == 0)

# Drop rows where ANY lab value is invalid
df = df[~invalid_mask.any(axis=1)].reset_index(drop=True)

# Sanity check
print("Dataset shape after dropping invalid lab rows:", df.shape)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.drop(['bnp_level',"ejection_fraction" ], axis=1)
```

```python
import numpy as np

lab_cols = [
    "hemoglobin",
    "total_leukocyte_count",
    "platelet_count",
    "glucose_level",
    "urea_level",
    "creatinine_level"
]

df[lab_cols] = df[lab_cols].replace(r"^\s*$", np.nan, regex=True)
```

```python
df[lab_cols] = df[lab_cols].apply(pd.to_numeric, errors="coerce")
```

```python
df[lab_cols].dtypes
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
df[lab_cols] = imputer.fit_transform(df[lab_cols])
```
